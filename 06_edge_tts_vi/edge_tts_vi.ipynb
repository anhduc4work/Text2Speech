# Edge TTS - Free Microsoft Neural Text-to-Speech for Vietnamese

## What is Edge TTS?

**Edge TTS** is a Python package that leverages Microsoft's neural text-to-speech engine — the same one powering the Microsoft Edge browser's built-in read-aloud feature.

### Why Edge TTS?

- **Completely free** — no API key, no subscription, no usage limits
- **No GPU required** — synthesis happens on Microsoft's servers, your machine just receives the audio
- **High quality** — uses the same neural voices as Azure Cognitive Services
- **Simple API** — just `pip install edge-tts` and you are ready

### Vietnamese Voices

Microsoft provides two neural voices for Vietnamese:

| Voice Name | Gender | Description |
|---|---|---|
| `vi-VN-HoaiMyNeural` | Female | Natural, clear female voice |
| `vi-VN-NamMinhNeural` | Male | Natural, clear male voice |

### How it works

Edge TTS sends text to Microsoft's TTS endpoint (the same one Edge browser uses) and streams back audio. Because the synthesis happens server-side, you need an internet connection but no local GPU or heavy model downloads.

```python
import edge_tts
import asyncio
import os
from IPython.display import Audio, display, Markdown

# Create output directory
os.makedirs("output", exist_ok=True)

print("edge-tts imported successfully!")
```

## 2. List All Vietnamese Voices

Let's query the available voices and filter for Vietnamese (`vi-VN`). This shows us exactly what voices are available and their properties.

```python
voices = await edge_tts.list_voices()

vi_voices = [v for v in voices if v["Locale"] == "vi-VN"]

print(f"Found {len(vi_voices)} Vietnamese voice(s):\n")
for v in vi_voices:
    print(f"  Name:           {v['ShortName']}")
    print(f"  Gender:         {v['Gender']}")
    print(f"  Friendly Name:  {v['FriendlyName']}")
    print(f"  Status:         {v['Status']}")
    print()
```

## 3. Basic Speech Generation

The core workflow is simple:
1. Create a `Communicate` object with your text and chosen voice
2. Call `save()` to write the audio to a file
3. Play it back with `IPython.display.Audio`

Let's generate the same sentence with both Vietnamese voices so you can compare them.

```python
text = "Xin chao! Toi la tro ly giong noi tieng Viet."

# Female voice: vi-VN-HoaiMyNeural
communicate = edge_tts.Communicate(text, "vi-VN-HoaiMyNeural")
await communicate.save("output/basic_female.mp3")

print("Female voice (vi-VN-HoaiMyNeural):")
display(Audio("output/basic_female.mp3", autoplay=False))
```

```python
# Male voice: vi-VN-NamMinhNeural
communicate = edge_tts.Communicate(text, "vi-VN-NamMinhNeural")
await communicate.save("output/basic_male.mp3")

print("Male voice (vi-VN-NamMinhNeural):")
display(Audio("output/basic_male.mp3", autoplay=False))
```

## 5. Speed, Pitch, and Volume Control

Edge TTS supports SSML-like prosody controls through the `Communicate` class parameters:

- **`rate`** — speaking speed, e.g. `"+50%"` (faster) or `"-30%"` (slower)
- **`pitch`** — voice pitch, e.g. `"+10Hz"` (higher) or `"-5Hz"` (lower)
- **`volume`** — audio volume, e.g. `"+20%"` (louder) or `"-20%"` (quieter)

These let you fine-tune the output without needing raw SSML.

```python
demo_text = "Day la mot cau vi du de kiem tra toc do, cao do, va am luong cua giong noi."

# Normal speed (baseline)
communicate = edge_tts.Communicate(demo_text, "vi-VN-HoaiMyNeural")
await communicate.save("output/prosody_normal.mp3")
print("Normal speed (default):")
display(Audio("output/prosody_normal.mp3", autoplay=False))
```

```python
# Slow speed (-30%)
communicate = edge_tts.Communicate(demo_text, "vi-VN-HoaiMyNeural", rate="-30%")
await communicate.save("output/prosody_slow.mp3")
print("Slow speed (rate=\"-30%\"):")
display(Audio("output/prosody_slow.mp3", autoplay=False))
```

```python
# Fast speed (+50%)
communicate = edge_tts.Communicate(demo_text, "vi-VN-HoaiMyNeural", rate="+50%")
await communicate.save("output/prosody_fast.mp3")
print("Fast speed (rate=\"+50%\"):")
display(Audio("output/prosody_fast.mp3", autoplay=False))
```

```python
# Higher pitch (+10Hz)
communicate = edge_tts.Communicate(demo_text, "vi-VN-HoaiMyNeural", pitch="+10Hz")
await communicate.save("output/prosody_high_pitch.mp3")
print("Higher pitch (pitch=\"+10Hz\"):")
display(Audio("output/prosody_high_pitch.mp3", autoplay=False))
```

```python
# Lower pitch (-10Hz)
communicate = edge_tts.Communicate(demo_text, "vi-VN-HoaiMyNeural", pitch="-10Hz")
await communicate.save("output/prosody_low_pitch.mp3")
print("Lower pitch (pitch=\"-10Hz\"):")
display(Audio("output/prosody_low_pitch.mp3", autoplay=False))
```

```python
# Combined: slow + low pitch (storytelling style)
communicate = edge_tts.Communicate(
    demo_text,
    "vi-VN-NamMinhNeural",
    rate="-20%",
    pitch="-5Hz",
    volume="+10%"
)
await communicate.save("output/prosody_combined.mp3")
print("Combined: rate=\"-20%\", pitch=\"-5Hz\", volume=\"+10%\" (male voice):")
display(Audio("output/prosody_combined.mp3", autoplay=False))
```

## 6. Batch Processing

A common use case is converting a list of sentences into individual audio files — for example, building a language learning dataset or generating voice prompts.

Below we process a list of Vietnamese sentences, saving each as a separate MP3 file.

```python
sentences = [
    "Xin chao, ban khoe khong?",
    "Hom nay thoi tiet rat dep.",
    "Toi thich an pho bo.",
    "Cam on ban rat nhieu.",
    "Hen gap lai ban sau.",
]

os.makedirs("output/batch", exist_ok=True)

voice = "vi-VN-HoaiMyNeural"

for i, sentence in enumerate(sentences, 1):
    filename = f"output/batch/sentence_{i:02d}.mp3"
    communicate = edge_tts.Communicate(sentence, voice)
    await communicate.save(filename)
    file_size = os.path.getsize(filename)
    print(f"[{i}/{len(sentences)}] Saved: {filename} ({file_size:,} bytes)")
    print(f"    Text: {sentence}")
    display(Audio(filename, autoplay=False))
    print()

print(f"\nBatch complete! Generated {len(sentences)} audio files.")
```

## 7. List Generated Files

Let's verify all the files we created.

```python
total_size = 0
file_count = 0

for root, dirs, files in os.walk("output"):
    for f in sorted(files):
        if f.endswith(".mp3"):
            path = os.path.join(root, f)
            size = os.path.getsize(path)
            total_size += size
            file_count += 1
            print(f"  {path:<45s} {size:>8,} bytes")

print(f"\nTotal: {file_count} files, {total_size:,} bytes ({total_size/1024:.1f} KB)")
```

## Key Takeaways

### Pros

- **Free with no limits** — no API key, no signup, no billing
- **High quality neural voices** — same engine as Azure Cognitive Services
- **No GPU needed** — synthesis runs on Microsoft servers; works on any machine
- **Simple API** — just a few lines of Python to generate speech
- **Prosody control** — adjust speed, pitch, and volume easily

### Cons

- **Requires internet** — no offline mode; depends on Microsoft's servers
- **No voice cloning** — you cannot train or fine-tune custom voices
- **Fixed voice set** — limited to the voices Microsoft provides (2 for Vietnamese)
- **No emotion/style control** — unlike Azure's full SSML, style tags are not supported
- **Unofficial API** — Microsoft could change or restrict access at any time

### When to use Edge TTS

Edge TTS is ideal for prototyping, personal projects, and cases where you need quick, free, high-quality Vietnamese speech synthesis without setting up infrastructure. For production systems requiring SLAs, custom voices, or offline capability, consider Azure Cognitive Services or other dedicated TTS solutions.
